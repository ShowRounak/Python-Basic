## Dataset Description:

- brand_name: Brand Name of Smartphones
- model: Speicific model name of a smartphone
- price: Price of a smartphone
- rating: Rating of the phone
- has_5g: If the phone has 5G connectivity
- has_nfc: if the phone has NFC feature
- has_ir_blaster: if the phone has IR blaster
- processor_brand: The brand/manufacturer of the processor used in a smartphone
- num_cores: Number of cores in a processor
- processor_speed: Speed of the processor in MHz
- battery_capacity: Battery capacity in mAh
- fast_charging_available: iF the phone has fast_charging support
- fast_charging: How much fast charging in Watt
- ram_capacity: ram_capacity of a phone
- internal_memory: internal_memory of a smartphone
- screen_size: Screen size of a phone
- refresh_rate: Refresh rate of the screen
- resolution: Resolution of the screen
- num_rear_cameras: Number of Rear cameras
- num_front_cameras: Number of front cameras
- os: OS of the phone
- primary_camera_rear: Megapixel count of primary camera sensor in rear
- primary_camera_front: Megapixel count of primary camera sensor in front
- extended_memory_available: If the phones support Micro SD card
- extended_upto: How much storage upgrade is supported

```python
num_cols = df.select_dtypes(include=['float64','int64']).iloc[:,[3,4,6,9,13,14,16]].columns
```

```python
for col in num_cols:
    plot_graphs(col)
```

## Bi-variate analysis

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df,x='brand_name',y='price')
plt.xticks(rotation='vertical')
plt.show()
```

```python
## lets check brands those have atleast 10 phones

x = df.groupby('brand_name').count()['model'].sort_values(ascending=False)
```

```python
brand_more_than_ten_phone = df[df['brand_name'].isin(x[x>10].index)]
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=brand_more_than_ten_phone,x='brand_name',y='price')
plt.xticks(rotation='vertical')
plt.show()
```

### Observation:
1. Apple and Huawei phones are usually high-priced.
2. Avg price of Samsung and Oneplus phones are 30K-45K INR.
3. Tecno, Infinix, Poco are generally make budget phones.
4. Brands which have less than 10 phones, we can merge them as 'Others'
