### So, by default pandas considers the first row of the file as column names.

### If any dataset doesnot have column names, then pandas will set the first row/record as the column name.

```python
# for example

pd.read_csv('taxonomy.csv')
```

But we can add our own column names if we want.

pd.read_csv() takes many arguments. Two of them are called header and names. If we set header = None, then it won't take the very first row as column name.


```python
pd.read_csv('taxonomy.csv', header=None)
```

If we want to add own custom column name, then we have to pass these names inside a list as names arguments/

```python
pd.read_csv('taxonomy.csv', names=['col1','col2', 'col3','col4'])
```

In the above example, there are 4 columns. What if we add 5 column names

```python
pd.read_csv('taxonomy.csv', names=['col1','col2', 'col3','col4','col5'])
```

#### How to read other sheets from an excel file

```python
pd.read_excel('LUSID Excel - Manage Orders.xlsx', sheet_name= 'Allocations')
```

```python
df =pd.read_excel('LUSID Excel - Manage Orders.xlsx')
```

How to get all sheetnames in a excel file

```python
df1 = pd.ExcelFile('LUSID Excel - Manage Orders.xlsx')
```

```python
df1.sheet_names
```

```python
for i in df1.sheet_names:
    pd.read_excel('LUSID Excel - Manage Orders.xlsx', sheet_name=i)
```
