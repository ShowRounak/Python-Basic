```python
# From a cabin column separate text and numeric value

df2 = df[df['Cabin'].isnull() == False]

```

```python
alpha = []

for i in df2['Cabin']:
    for j in i:
        if j.isalpha():
            #print(j)
            alpha.append(j)
print(alpha)
```

```python
num = []

for i in df2['Cabin']:
    for j in i:
        if j.isnumeric():
            #print(j)
            num.append(j)
    
print(num)
```
