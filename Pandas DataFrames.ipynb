```python
def track_record(team1, team2):
    """ Function to check track record of two IPL teams"""
    home_record = ipl[(ipl['Team1'] == team1) & (ipl['Team2'] == team2)]['WinningTeam'].value_counts()
    away_record = ipl[(ipl['Team1'] == team2) & (ipl['Team2'] == team1)]['WinningTeam'].value_counts()
    if team1 in home_record and team1 in away_record and team2 in home_record and team2 in away_record:
        team1_won = home_record[team1] + away_record[team1]
        team2_won = home_record[team2] + away_record[team2]
        print(f"{team1} won {team1_won} and {team2} won {team2_won}")
    elif team1 not in home_record and team1 not in away_record:
        team2_won = home_record[team2] + away_record[team2]
        print(f"{team1} won 0 times and {team2} won {team2_won} times")
    elif team2 not in home_record and team2 not in away_record:
        team1_won = home_record[team1] + away_record[team1]
        print(f"{team2} won 0 times and {team1} won {team1_won} times")
```

```python
track_record(team2='Rajasthan Royals',team1='Royal Challengers Bangalore')
```

```python
track_record(team2='Chennai Super Kings',team1='Kolkata Knight Riders')
```

```python
track_record('Mumbai Indians', 'Kolkata Knight Riders')
```

```python
track_record('Rajasthan Royals', 'Gujarat Titans')
```
